График рассеяния (`scatter plot`) является полезным инструментом для визуализации взаимосвязи между двумя переменными. Однако в случае, когда одна из переменных является категориальной (например, "Pclass" - класс пассажира), использование графика рассеяния не является наилучшим выбором.

Проблема заключается в том, что график рассеяния подразумевает, что значения переменной на оси x распределены непрерывно, а не дискретно. В данном случае, "Pclass" имеет всего несколько категорий (1, 2, 3), и точки на графике будут сосредоточены вокруг отдельных дискретных значений, что делает график рассеяния менее информативным и не очень показательным.

Вишенка на торте: рисуем несколько графиков на одной фигуре.

```python
plt.figure(figsize=(15, 5), dpi=300)

plt.subplot(1, 2, 1)
plt.title("1")
plt.hist(df["Age"])
plt.ylabel("Count")
plt.xlabel("Age")

plt.subplot(1, 2, 2)
plt.title("2")
plt.hist(df["Fare"])
plt.xlabel("Fare")

plt.show()
```

В способе выше мы должны каждый раз переключаться между тем, где мы рисуем, при помощи `plt.subplot(n_rows, n_cols, idx)`. Часто это не очень удобно, поэтому давайте посмотрим на другой способ взаимодействия с подграфиками. Этот способ позволит, например, удобно итерироваться в цикле по графикам и отрисовывать что-то.

```python
fig, axs = plt.subplots(1, 2, figsize=(15, 5), dpi=300)

axs[0].set_title("1")
axs[0].hist(df["Age"])
axs[0].set_ylabel("Count")
axs[0].set_xlabel("Age")

axs[1].set_title("2")
axs[1].hist(df["Fare"])
axs[1].set_xlabel("Fare")

plt.show()
```

Так мы нарисовали два графика рядом друг с другом, используя функцию `plt.subplots()`, которая возвращает объекты `fig` (рисунок) и `axs` (ось или оси графиков).


1. `fig, axs = plt.subplots(1, 2, figsize=(15, 5), dpi=300)`: Эта строка создает один рисунок (`fig`) и две оси графиков (`axs`) в одной строке и два столбца (1 строка и 2 столбца), чтобы разместить два графика рядом. Параметр `figsize` устанавливает размер рисунка (ширина, высота) в дюймах, а `dpi` (dots per inch) устанавливает количество точек на дюйм, что влияет на разрешение графиков.

2. `axs[0].set_title("1")`: Устанавливает заголовок для первого графика.

3. `axs[0].hist(df["Age"])`: Создает гистограмму распределения возраста пассажиров и отображает ее на первом графике.

4. `axs[0].set_ylabel("Count")`: Устанавливает подпись оси y для первого графика.

5. `axs[0].set_xlabel("Age")`: Устанавливает подпись оси x для первого графика.

6. `axs[1].set_title("2")`: Устанавливает заголовок для второго графика.

7. `axs[1].hist(df["Fare"])`: Создает гистограмму распределения цен на билеты и отображает ее на втором графике.

8. `axs[1].set_xlabel("Fare")`: Устанавливает подпись оси x для второго графика.

9. `plt.show()`: Отображает созданные графики.

Тем самым, анализируя график barplot, мы можем сравнить средние значения выживаемости для мужчин и женщин и оценить статистическую значимость этой разницы, учитывая доверительные интервалы.

Ещё один полезный вид визуализации — ящик с усами. Такой вид графиков позволяет визуально оценить моду и разброс распределения признака. Посмотрим на ящик с усами, отражающий распределение пассажиров по полу и возрасту.

```python
sns.catplot(x="Sex", y="Age", data=Data, kind="box")  # box plot (box-and-whiskers-plot)
plt.show()
```

Ящик с усами (box plot) - это графическое представление распределения данных через квартили. Он позволяет оценить центральную тенденцию, разброс и форму распределения данных, а также выявить наличие выбросов.

Вот как интерпретировать box plot:

- **Ящик (Box)**: Ящик показывает интерквартильный размах (IQR), который представляет собой разницу между верхним (75-й процентиль) и нижним (25-й процентиль) квартилями. Средний (50-й процентиль) обычно отмечен линией внутри ящика. Чем больше высота ящика, тем больше разброс значений внутри него.
  
- **Усы (Whiskers)**: Усы представляют собой линии, которые расширяются от верхнего и нижнего квартилей до пределов наблюдений, не считая выбросы. Длина усов может варьироваться в зависимости от выбранной методологии для определения выбросов (например, 1.5 межквартильных размаха).

- **Точки (Outliers)**: Отдельные точки представляют собой выбросы - значения, находящиеся за пределами усов. Они могут быть важными для дальнейшего анализа, так как указывают на потенциально необычные или аномальные наблюдения.

Используя `sns.catplot()` с параметром `kind="box"`, мы создаем box plot для визуализации распределения переменной ("Age") в зависимости от категориальной переменной ("Sex"). В данном случае, box plot позволяет сравнить распределение возраста между мужчинами и женщинами. Вертикальная линия внутри каждого ящика показывает медианное значение возраста, а длина ящика и усов позволяют судить о разбросе возраста в каждой группе.

2. Для каждой уникальной категории в этом столбце создается новый бинарный столбец. Если у нас есть, например, три уникальные категории в столбце "Embarked", мы получим три новых бинарных столбца.

Зачастую признак может быть зашумлен или может содержать лишнюю (слишком детализированную, ведущую к переобучению) информацию. Возможным решением этой проблемы служит бинаризация признака.

Бинаризуем признаки `Age` и `Fare`.

* `pd.cut` — разбиение целочисленных данных на несколько интервалов по квантилям
* `pd.qcut` — разбиение числовых (необязательно целочисленных) данных на несколько интервалов по квантилям

```python
Data["AgeBin"] = pd.cut(Data["Age"].astype(int), 5)
Data["AgeBin"].head()
```

Бинаризация признаков может быть полезным методом предобработки данных, особенно когда нам нужно сократить шум или уменьшить количество информации в признаке.

Для бинаризации признаков `Age` и `Fare` мы можем использовать `pd.cut()` или `pd.qcut()` для разбиения числовых данных на интервалы.

 **pd.cut()**: Этот метод разбивает данные на интервалы, создавая категориальную переменную, в которой каждое значение будет указывать на интервал, в которое оно попадает. Мы можем задать количество интервалов с помощью параметра `bins`.



Теперь переведём полученные интервалы в числа, используя `LabelEncoder`.

```python
from sklearn.preprocessing import LabelEncoder

label = LabelEncoder()
Data["AgeBin_Code"] = label.fit_transform(Data["AgeBin"])
Data[["Age", "AgeBin", "AgeBin_Code"]].head()
```
